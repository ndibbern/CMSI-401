```python
import pandas as pd
import arff
import numpy as np
import scipy as sp
import sklearn
```

```python
df = pd.read_csv('tao-all2.txt', sep=" ", header=None)
df.columns = ['Observation','Year', 'Month', 'Day', 'Date', 'Latitude', 'Longitude', 'Zonal_Winds', 'Meridional_Winds', 'Humidity', 'Air_Temp', 'Sea_Surface_Temp']
df.head()
```

```python
del df['Observation']
del df['Date']
df.head()
```

```python
df['Humidity'] = df.apply(lambda x: float(x.Humidity) if x.Humidity != '.' and x.Humidity != None else None, axis=1)
df['Zonal_Winds'] = df.apply(lambda x: float(x.Zonal_Winds) if x.Zonal_Winds != '.' and x.Zonal_Winds != None else None, axis=1)
df['Meridional_Winds'] = df.apply(lambda x: float(x.Meridional_Winds) if x.Meridional_Winds != '.' and x.Meridional_Winds != None else None, axis=1)
df['Air_Temp'] = df.apply(lambda x: float(x.Air_Temp) if x.Air_Temp != '.' and x.Air_Temp != None else None, axis=1)
df['Sea_Surface_Temp'] = df.apply(lambda x: float(x.Sea_Surface_Temp) if x.Sea_Surface_Temp != '.' and x.Sea_Surface_Temp != None else None, axis=1)
df.head()
```

```python
df['Humidity'].fillna((df['Humidity'].mean()), inplace=True)
df['Zonal_Winds'].fillna((df['Zonal_Winds'].mean()), inplace=True)
df['Meridional_Winds'].fillna((df['Meridional_Winds'].mean()), inplace=True)
df['Air_Temp'].fillna((df['Air_Temp'].mean()), inplace=True)
df['Sea_Surface_Temp'].fillna((df['Sea_Surface_Temp'].mean()), inplace=True)
df.head()
```

```python
wanted = ['Zonal_Winds', 'Meridional_Winds', 'Humidity', 'Air_Temp','Sea_Surface_Temp' ]
df[wanted].head()
```

```python
df.to_csv('tao-cleaned.csv')
```
